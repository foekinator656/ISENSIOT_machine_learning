```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
afbeelding = cv.imread('kodak dataset/kodim23.png')
# convert to RGB
grijs = cv.cvtColor(afbeelding, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(grijs,cmap='gray')
plt.show()
```

```python
f_g = grijs.flatten()
```

```python
plt.hist(f_g,bins=50)
plt.show()
```

```python
# De drempelwaarde
newimage = grijs>150
```

```python
# Je ziet dat door de drempelwaarde de afbeelding wordt
# geconverteerd naar True of False
newimage
```

```python
# Het resultaat van de thresholding is een binaire image
plt.imshow(newimage,cmap='gray')
plt.show()
```
