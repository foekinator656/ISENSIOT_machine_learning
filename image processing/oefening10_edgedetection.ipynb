```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
im1 = plt.imread('kodak dataset/kodim23.png')
```

```python
plt.imshow(im1)
plt.show()
```

```python
im1.shape
```

```python
red = im1[:,:,0]
green = im1[:,:,1]
blue = im1[:,:,2]
```

```python
gray = 0.299*red + 0.587*green + 0.114*blue
```

```python
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
# Dit zijn de edge detection kernels
Gx = np.matrix([[1,2,1],[0,0,0],[-1,-2,-1]])
Gy = np.matrix.transpose(Gx)
print(Gx)
print()
print(Gy)
```

```python
edgex = cv2.filter2D(gray,-1,Gx)
edgey = cv2.filter2D(gray,-1,Gy)
plt.imshow(edgex,cmap='gray')
plt.show()
```

```python
# Voeg de twee kernel resultaten samen. 
# Dit is gewoon de stelling van Pythagoras
w = np.sqrt(np.square(edgex)+np.square(edgey))
```

```python
plt.imshow(w,cmap='gray')
plt.show()
```

```python
# Gebruik hier thresholding om een de edges duidelijker te maken. 
# Daarvoor plotten we eerst een histogram om de threshold
# te bepalen
plt.hist(w.flatten(),bins=50)
plt.show()
```

```python
plt.imshow(w>0.5,cmap='gray')
plt.show()
```
