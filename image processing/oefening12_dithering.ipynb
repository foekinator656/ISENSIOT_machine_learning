### Dithering
Hier maken we een ruis afbeelding aan. Deze tellen we op bij de originele afbeelding. Dit soort ervoor dat gladde banden in de afbeelding verdwijnen. En er komt meer detail tevoorschijn. 

```python
# Ruis afbeelding
dithering = np.random.rand(512, 768)*70
```

```python
# We plotten de ruis om te zien dat het ook echt ruis is. 
plt.imshow(dithering,cmap='gray')
plt.show()
```

```python
# De drempelwaarde met ruis bij elkaar opgeteld
newimage = grijs + dithering>150
```

```python
# Je ziet dat door de drempelwaarde de afbeelding wordt
# geconverteerd naar True of False
newimage
```

```python
# Het resultaat van de thresholding plus dithering
# is een binaire afbeelding waarbij details redelijk bewaard blijven
plt.imshow(newimage,cmap='gray')
plt.show()
```
