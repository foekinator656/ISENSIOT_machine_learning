```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
im1 = plt.imread('kodak dataset/kodim23.png')
```

```python
plt.imshow(im1)
plt.show()
```

```python
im1.shape
```

```python
mask = np.ones([512, 768])
```

```python
# Hier selecteren we een gebied dat we zwart maken (0)
mask[100:400,300:500]=0
```

```python
plt.imshow(mask,cmap='gray')
plt.show()
```

```python
# inverteer het masker en maak het binair
mask = mask!=1
```

```python
# Elk afzonderlijk kleur kanaal 
# gaan we met het masker vermenigvuldigen
im1[:,:,0] = im1[:,:,0]*mask
im1[:,:,1] = im1[:,:,1]*mask
im1[:,:,2] = im1[:,:,2]*mask
```

```python
# Het resultaat is een mask over het originele image
plt.imshow(im1)
plt.show()
```
