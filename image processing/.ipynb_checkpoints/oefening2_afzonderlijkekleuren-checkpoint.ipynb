```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
afbeelding = cv.imread('kodak dataset/RGB_kleuren.jpg')
# converteer naar RGB
afbeelding = cv.cvtColor(afbeelding, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(afbeelding)
plt.show()
```

```python
# de grootte van de afbeelding
afbeelding.shape
```

```python
# de afzonderlijke kleurkanalen (rgb). 
# Kijk of je elk kanaal kan visualiseren
rood  = afbeelding[:,:,0]
groen = afbeelding[:,:,1]
blauw = afbeelding[:,:,2]
```

```python
# Verander hier 'rood' door 'groen' of 'blauw'.
# Dan zie je het vierkant of driehoek. 
plt.imshow(rood,cmap='gray')
plt.show()
```

```python
# Maak een lege image aan. 
w = np.zeros([240,320,3])
w.shape
```

```python
# probeer zelf de kleuren te maken. 
# Matplotlib gebruikt genormaliseerde kleur waarden tussen 0 en 1. 
# Dus 255 is in matplotlib 1. 
w[:,:,0]=0 # rood
w[:,:,1]=1 # groen
w[:,:,2]=0 # blauw
```

```python
plt.imshow(w)
plt.show()
```
