```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
im1 = plt.imread('kodak dataset/lenna.png')
#im1 = cv2.imread('kodak dataset/aa.jpg')
```

```python
plt.imshow(im1)
plt.show()
```

```python
im1.shape
```

```python
red = im1[:,:,0]
green = im1[:,:,1]
blue = im1[:,:,2]
```

```python
gray = 0.299*red + 0.587*green + 0.114*blue
```

```python
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
Gx = np.matrix([[1,2,1],[0,0,0],[-1,-2,-1]])
Gy = np.matrix.transpose(Gx)
print(Gx)
print(Gy)
```

```python
edgex = cv2.filter2D(gray,-1,Gx)
edgey = cv2.filter2D(gray,-1,Gy)
plt.imshow(edgex,cmap='gray')
plt.show()
```

```python
w = np.sqrt(np.square(edgex)+np.square(edgey))
```

```python
plt.imshow(w,cmap='gray')
plt.show()
```

```python
plt.hist(w.flatten(),bins=50)
plt.show()
```

```python
plt.imshow(w>0.4,cmap='gray')
plt.show()
```
