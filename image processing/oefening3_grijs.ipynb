```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
afbeelding = cv.imread('kodak dataset/kodim23.png')
# convert to RGB
afbeelding = cv.cvtColor(afbeelding, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(afbeelding)
plt.show()
```

```python
red = afbeelding[:,:,0]
print(red.shape)
print(red[200,600])
green = afbeelding[:,:,1]
blue = afbeelding[:,:,2]
```

```python
gray = 0.299*red + 0.587*green + 0.114*blue
```

```python
plt.imshow(gray,cmap='gray')
plt.show()
```
