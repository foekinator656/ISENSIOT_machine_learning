```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
afbeelding = cv.imread('kodak dataset/RGB_kleuren.jpg')
# convert to RGB
afbeelding = cv.cvtColor(afbeelding, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(afbeelding)
plt.show()
```

```python
afbeelding.shape
```

```python
# de afzonderlijke kleurkanalen (rgb)
rood  = afbeelding[:,:,0]
groen = afbeelding[:,:,1]
blauw = afbeelding[:,:,2]
```

```python
flatten_groen = groen.flatten()
flatten_groen.shape
```

```python
# Hier zie je dat alles overwegend zwart is (0) 
# En de andere kleur is 255. 
plt.hist(flatten_groen,bins=50)
plt.show()
```
